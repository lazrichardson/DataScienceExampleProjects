## Summary
This project takes an input image file and slices it into equally sized sections. Then, it runs the MASS alorithm on each piece, comparing it to an input reference image. The goal is to use the algorithm to find the are of the image that contains the reference image.

The MASS algorithm is Mueen's Algorithm for Similarity Search. It is a very fast similarity search for time series subsequences. It uses euclidian distance and correlation coefficient to find occurences of similar motifs either within one timeseries, across multiple timeseries, or to find one motif within another timeseries.
